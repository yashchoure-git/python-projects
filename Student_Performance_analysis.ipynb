```python
#load file
student_file = 'student_scores.xlsx'
df = pd.read_excel(student_file)

```

```python
#examine file
df.head()
```

```python
#calculate avarage score
df["Avarage"] = df.iloc[:,1:].mean(axis=1)

df.head()
```

```python
#Find Top performer
Top_performer = df.loc[df["Avarage"].idxmax(),"Student"]
print(Top_performer)
```

```python
#subject wise averages
subject_wiseaverage = df.iloc[:,1:4].mean()
print(subject_wiseaverage)
```
