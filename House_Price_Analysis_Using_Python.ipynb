**House Price Analysis Using Python**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from google.colab import files
uploaded=files.upload()
```

```python
house_data=pd.read_csv('house_price_data.csv')
house_data.head()
```

1) House Price Distribution

```python
sns.distplot(house_data['price'],color='green',bins=10)
plt.show()
```

2)Price vs. Square Footage

```python
sns.scatterplot(x='sqft',y='price',data=house_data,hue='price')
plt.show()
```

3)Features relate to price.

```python
matrix=house_data.corr()
plt.figure(figsize=(8,6))

sns.heatmap(matrix,annot=True,cmap='coolwarm',linewidth=0.5)
plt.show()
```

4)House Price By no. of Beadrooms

```python
sns.boxplot(x='bedrooms',y='price',data=house_data,palette='Set1')
plt.show()
```
